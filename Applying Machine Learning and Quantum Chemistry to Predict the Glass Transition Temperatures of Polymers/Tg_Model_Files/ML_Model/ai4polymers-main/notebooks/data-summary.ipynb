# Data distribution

```python
n_bins=50

plt.figure(figsize=(9,6), dpi=300, facecolor='white')
plt.subplots_adjust(hspace=0.4)

plt.subplot(2,3,1)
plt.title('$T_g$')
plt.hist(df_tg.iloc[:,1], bins=n_bins, color='purple')
plt.ylabel('Monomer count (#)')
plt.xlabel('$\\degree C$')

plt.subplot(2,3,2)
plt.title('$T_m$')
plt.hist(df_tm.iloc[:,1], bins=n_bins, color='gold')
plt.xlabel('$\\degree C$')

plt.subplot(2,3,3)
plt.title(r'$\rho$')
plt.hist(df_d.loc[df_d.iloc[:,1]<200].iloc[:,1], bins=n_bins, color='skyblue')
plt.xlabel('$g/cm^3$')

plt.subplot(2,3,4)
#plt.title('$T_g$')
plt.hist(np.log10(df_tg.iloc[:,1]+273.15), bins=n_bins, color='purple')
plt.ylabel('Monomer count (#)')
plt.xlabel('$log_{10}(K)$')

plt.subplot(2,3,5)
#plt.title('$T_m$')
plt.hist(np.log10(df_tm.iloc[:,1]+273.15), bins=n_bins, color='gold')
plt.xlabel('$log_{10}(K)$')

plt.subplot(2,3,6)
#plt.title(r'$\rho$')
plt.hist(np.log10(df_d.loc[df_d.iloc[:,1]<200].iloc[:,1]), bins=n_bins, color='skyblue')
plt.xlabel('$log_{10}(g/cm^3$)')
```
