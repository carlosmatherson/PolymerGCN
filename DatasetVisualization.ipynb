# Dataset Visualization

Import

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Curated Liu PolyInfo dataset Visualization

```python
mydata = pd.read_csv("https://raw.githubusercontent.com/carlosmatherson/PolymerGCN/main/SMILESwithAllProps.csv")
mytg = mydata["Tg"]
mymt = mydata["Mt"]
myro = mydata["Density"]
```

```python
fig, axs = plt.subplots(2,2)
plt.subplots_adjust(bottom=0.25, right=1.5, top=2)
axs[0,0].hist(mytg, bins=50, color="orange")
axs[0,0].set_xlabel(r'${\rm T_{g}}$ ${\rm(^{o}C)}$', fontsize=14)
axs[0,0].set_ylabel(r'Counts', fontsize=14)
axs[0,1].hist(mymt, bins=50, color="orange")
axs[0,1].set_xlabel(r'${\rm T_{m}}$ ${\rm(^{o}C)}$', fontsize=14)
axs[0,1].set_ylabel(r'Counts', fontsize=14)
axs[1,0].hist(myro, bins=50, color="orange")
axs[1,0].set_xlabel(r'${\rm \rho}$ ${\rm{(g/cm^3)}}$', fontsize=14)
axs[1,0].set_ylabel(r'Counts', fontsize=14) #_ = to silence pyplot text output
axs[-1, -1].axis('off')
plt.show()
```

GREA Dataset Visualization

```python
tgRead = pd.read_csv("https://raw.githubusercontent.com/liugangcode/GREA/main/data/tg_prop/raw/tg_raw.csv")
mtRead = pd.read_csv("https://raw.githubusercontent.com/liugangcode/GREA/main/data/mt_prop/raw/mt_raw.csv")
dsRead = pd.read_csv("https://raw.githubusercontent.com/liugangcode/GREA/main/data/density_prop/raw/density_raw.csv")
dsRead = dsRead[dsRead["density"] < 10]
tgData = tgRead["tg"]
mtData = mtRead["mt"]
dsdata = dsRead["density"]
```

```python
fig, axs = plt.subplots(2,2)
plt.subplots_adjust(bottom=0.25, right=1.5, top=2)
axs[0,0].hist(tgData, bins=50, color="orange")
axs[0,0].set_xlabel(r'${\rm T_{g}}$ ${\rm(^{o}C)}$', fontsize=14)
axs[0,0].set_ylabel(r'Counts', fontsize=14)
axs[0,1].hist(mtData, bins=50, color="orange")
axs[0,1].set_xlabel(r'${\rm T_{m}}$ ${\rm(^{o}C)}$', fontsize=14)
axs[0,1].set_ylabel(r'Counts', fontsize=14)
axs[1,0].hist(dsdata, bins=50, color="orange")
axs[1,0].set_xlabel(r'${\rm \rho}$ ${\rm{(g/cm^3)}}$', fontsize=14)
axs[1,0].set_ylabel(r'Counts', fontsize=14) #_ = to silence pyplot text output
axs[-1, -1].axis('off')
plt.show()
```
